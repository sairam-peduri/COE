```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"data.csv")
data
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
obj=data.select_dtypes(include=['object']).columns
obj
```

```python
for i in obj:
    if(data[i].isna().sum()!=0):
        mval=data[i].mode()[0]
        data[i].replace(np.NaN,mval,inplace=True)
data.isna().sum()
```

```python
data[['date', 'StartDate', 'OpeningDate', 'Deadline']] = data[['date', 'StartDate', 'OpeningDate', 'Deadline']].apply(pd.to_datetime, errors='coerce')
data.dtypes
```
