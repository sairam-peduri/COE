```python
me=gt['AverageTemperature'].mean()
me
```

```python
import numpy as np
gt['AverageTemperature']=gt['AverageTemperature'].replace(np.NaN,me)
```

```python
gt
```

```python
mu=gt['AverageTemperatureUncertainty'].mean()
mu
```

```python
gt['AverageTemperatureUncertainty']=gt['AverageTemperatureUncertainty'].replace(np.NaN,mu)
```

```python
gt
```

```python
gt.dtypes
```

```python
gt['dt']=pd.to_datetime(gt['dt'])
gt.dtypes
```

```python
y=gt['dt'].dt.year
```

```python
y
```

```python
gt['year']=y
```

```python
gt
```

```python
annual_avg_temp=gt.groupby('year')['AverageTemperature'].mean().reset_index()
```

```python
annual_avg_temp
```

```python
import matplotlib.pyplot as py
py.plot(annual_avg_temp['year'],annual_avg_temp['AverageTemperature'])
py.show
```

```python
country_temp=gt.groupby('Country')['AverageTemperature'].mean().reset_index()
```

```python
country_temp
```

```python
cs=country_temp.sort_values(by='AverageTemperature',ascending=False).reset_index()
```

```python
cs
```

```python
ct=cs.head(10)
```

```python
ct
```

```python
py.bar(ct['Country'],ct['AverageTemperature'],color='y',edgecolor='black')
py.title('Top 10 countries with highest Temperature')
py.show
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as py
tips=sns.load_dataset('tips')
tips.head(5)
```

```python
sns.histplot(x='total_bill',data=tips,kde=True)
```

```python
sns.kdeplot(tips['total_bill'],color="blue",fill=True)
```

```python
sns.barplot(tips['total_bill'])
```

```python
sns.boxplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])
```

```python
sns.pairplot(data=tips,hue='sex',palette='deep')
```

```python
numeric_tips = tips.select_dtypes(include=['number'])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
py.show()
```
